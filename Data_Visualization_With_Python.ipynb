```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1,color="g",linestyle=":",linewidth = 2)
plt.plot(x,y2)
plt.title("Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="g",linestyle=":",linewidth = 2)
plt.title("Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid()
plt.show()
```
